```python
# outlier treatment for price
plt.boxplot(Datos.km_recorridos)
Q1 = Datos.km_recorridos.quantile(0.25)
Q3 = Datos.km_recorridos.quantile(0.75)
IQR = Q3 - Q1
Datos = Datos[(Datos.km_recorridos >= Q1 - 1.5*IQR) & (Datos.km_recorridos <= Q3 + 1.5*IQR)]
```

Visualización de variables categóricas

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'año', y = 'precio_de_venta', data = Datos)
plt.subplot(2,3,2)
sns.boxplot(x = 'asientos', y = 'precio_de_venta', data = Datos)
plt.subplot(2,3,3)
sns.boxplot(x = 'km_recorridos', y = 'precio_de_venta', data = Datos)

plt.show()
```
